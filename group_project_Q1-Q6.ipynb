Q1. Load the data set and do data processing

```python
#basic information of the data set
print()
print("GSM example:")
for gsm_name, gsm in gse.gsms.items():
    print("Name: ", gsm_name)
    print("Metadata:",)
    for key, value in gsm.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print ("Table data:",)
    print (gsm.table.head())
    break
```

```python
#gene information from Gene Expression Omnibus Platform (GPL)
print()    
print("GPL example:")    
for gpl_name, gpl in gse.gpls.items():    
    print("Name: ", gpl_name)    
    print("Metadata:",)    
    for key, value in gpl.metadata.items():    
        print(" - %s : %s" % (key, ", ".join(value)))    
    print("Table data:",)    
    print(gpl.table.head())    
    break   
gpl.table.head()
```

```python
gpl.table["Gene Symbol"].isnull().any()
gpl.table.columns
```

```python
#Obatian gene information table: extract Identifier and Gene symbol with gene id
metadata = pd.DataFrame(gpl.table, columns=['ID', 'Gene Symbol', 'ENTREZ_GENE_ID'])
metadata.dropna(inplace=True)
metadata
```

```python
#extract gene expression matrix from gse, which includes gene expression of 58 samples
expr = pd.DataFrame(gse.pivot_samples('VALUE'))
expr.head()
```

```python
#merge expression matrix with gene information table
gene_id_df = pd.merge(metadata,expr, left_on= "ID", right_on="ID_REF")
gene_id_df
```

```python
#each gene may have mutiple ID, only keep the ID with the highest gene expression. 
#Using mean gene expression value to sort
int_cols = gene_id_df.select_dtypes(include='float')
gene_id_df["mean"] = int_cols.mean(axis=1)

# drop duplicated gene symbol, keep the first occurance of each gene symbol to keep highest mean expression
gene_id_df = gene_id_df.sort_values(by=["mean"],ascending=False)
gene_id_no_dupes = gene_id_df.drop_duplicates("Gene Symbol")

# only keep gene symbol and gene expression value of 58 samples
expr_df = gene_id_no_dupes.drop(["ID", "mean", "ENTREZ_GENE_ID"], axis= 1)

# set Gene Symbol name as index
expr_df = expr_df.set_index(["Gene Symbol"])
expr_df
```

```python
alldata=expr_df.T
alldata
```

After above data processing, we get the data set of 58 samples gene expression corresponding to each gene symbol.

Q2. Perform clustering using k = 2 

```python
kmeans= KMeans(n_clusters=2,random_state=66)
kmeans
```

Q3. create a DataFrame that contains the 58 sample labels as one column, the cluster number as another column and the subtype given in the dataset (adenocarcinoma or squamous cell carcinoma) as another column.

```python
#use the kmeans model
kmeans_alldata = kmeans.fit(alldata)
```

```python
#new dataframe, contains the 58 sample labels as one column, 
#the cluster number as another column and the subtype given in the dataset 
index_data = alldata.index

#Sample label
sample_label_df = pd.DataFrame(index_data, columns=['Sample Label'])
sample_label_df = sample_label_df.set_index('Sample Label')

#2.Cluster Number
clust_labels = kmeans_alldata.labels_
sample_label_df['Cluster Number'] = clust_labels

# #3.Subtype
subtype_series = gse.phenotype_data["title"].str.split('_').str[1]
subtype_series_df = subtype_series.to_frame('Subtype')
sample_clust_subtype = sample_label_df.merge(subtype_series_df,left_index=True, right_index=True)
sample_clust_subtype
```

Q4.Compute and display the accuracy of this clustering.  

This is assuming that genes in cluster 1 have AC subtype and a genes in cluster 0 have SCC subtype
<br>The accuracy computes percentage of correct assignments to cluster 1 and 0


TP: cluster number = 1 & subtype = AC  
FN: cluster number = 0 & subtype = SCC  
TN: cluster number = 1 & subtype = SCC  
FP: cluster number = 0 & subtype = AC

```python
TP_alldata = sum((sample_clust_subtype['Cluster Number'] == 1) & (sample_clust_subtype['Subtype'] == 'AC'))
FN_alldata = sum((sample_clust_subtype['Cluster Number'] == 0) & (sample_clust_subtype['Subtype'] == 'SCC'))
TN_alldata = sum((sample_clust_subtype['Cluster Number'] == 1) & (sample_clust_subtype['Subtype'] == 'SCC'))
FP_alldata = sum((sample_clust_subtype['Cluster Number'] == 0) & (sample_clust_subtype['Subtype'] == 'AC'))
#also, TP+FN+TN+FP = 58
```

Accuracy = (TP+NF)/(TP+NF+TF+NP)


```python
accuracy_alldata = (TP_alldata+FN_alldata)/(TP_alldata+FN_alldata+TN_alldata+FP_alldata)
print("Accuracy:", accuracy_alldata)
```

Q5. Fit the model on half of the data (balanced: 20 AD and 9 SCC) and assign a label to the clusters (AD or SCC), then use the model to predict the cluster/label for the other half of the data.

```python
from sklearn.model_selection import train_test_split
```

```python
features = alldata
subtype_labels = sample_clust_subtype["Subtype"]
```

```python
# splitting all data into 50% test and 50% training sets
# stratify will split the subtypes AC (20) and SCC (9) evenly 
x_train, x_test, y_train, y_test = train_test_split(features, subtype_labels, test_size=0.5, random_state=66, stratify=subtype_labels)
```

```python
# fit the model to training set
kmeans_train = kmeans.fit(x_train)
```

```python
# view cluster labels for training set
kmeans_train_labels = kmeans_train.labels_
kmeans_train_labels
```

```python
# predict cluster labels for test set
kmeans_predict = kmeans.predict(x_test)
kmeans_predict
```

Q6. Compute and display the accuracy of the model for the prediction results (testing data)

```python
# create df with sample labels of test set, predicted cluster number and original subtype info
index_test_data = y_test.index

# Sample label
test_sample_df = pd.DataFrame(index_test_data, columns=['Sample Label'])
test_sample_df = test_sample_df.set_index('Sample Label')

# Cluster Number
test_sample_df['Cluster Number'] = kmeans_predict

# Subtype
subtype_test_df = y_test.to_frame('Subtype')
test_predicted_df = test_sample_df.merge(subtype_test_df,left_index=True, right_index=True)
test_predicted_df
```

Assuming that genes in cluster 1 have AC subtype and a genes in cluster 0 have SCC subtype

TP: cluster number = 1 & subtype = AC  
FN: cluster number = 0 & subtype = SCC  
TN: cluster number = 1 & subtype = SCC  
FP: cluster number = 0 & subtype = AC

```python
TP_counts = sum((test_predicted_df['Cluster Number'] == 1) & (test_predicted_df['Subtype'] == 'AC'))
FN_counts = sum((test_predicted_df['Cluster Number'] == 0) & (test_predicted_df['Subtype'] == 'SCC'))
TN_counts = sum((test_predicted_df['Cluster Number'] == 1) & (test_predicted_df['Subtype'] == 'SCC'))
FP_counts = sum((test_predicted_df['Cluster Number'] == 0) & (test_predicted_df['Subtype'] == 'AC'))
#also, TP+FN+TN+FP = 29
```

```python
Accuracy = (TP_counts+FN_counts)/(TP_counts+FN_counts+TN_counts+FP_counts)
print("Accuracy:", Accuracy)
```
