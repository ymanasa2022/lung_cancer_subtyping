```python
#Perform clustering - you can use the sklearn.cluster module for this task.
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
```

```python
#注释信息
gpl.table.head()
```

```python

gpl.table["Gene Symbol"].isnull().any()
gpl.table.columns
```

```python
#提取探针及基因名
metadata = pd.DataFrame(gpl.table, columns=['ID', 'Gene Symbol', 'ENTREZ_GENE_ID'])
metadata
```

```python
metadata.dropna(inplace=True)
metadata
```

```python
#提取表达矩阵
expr = pd.DataFrame(gse.pivot_samples('VALUE'))
expr.head()
```

```python

df = pd.merge(metadata,expr, left_on= "ID", right_on="ID_REF")
df
```

```python
#同一个基因对应多个探针，只保留表达量最大的探针
df["mean"] = df.mean(axis="columns")
df = df.sort_values(by=["mean"],ascending=False)
#去重
df = df.drop_duplicates("Gene Symbol")
df
```

```python

#去除列

df = df.drop("mean",1)
df = df.drop("ID",1)
df = df.drop("ENTREZ_GENE_ID",1)
df
```

```python
#将gene symbol设为行名
df = df.set_index(["Gene Symbol"])
df
```

```python
model= KMeans(n_clusters=2)
model
```

```python
cdata=df.T
cdata.head()
```

```python

model.fit(cdata)
```

```python
print( model.cluster_centers_, '\n')
print( model.labels_, '\n' )
print( model.inertia_, '\n')
```
